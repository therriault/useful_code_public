# Visualizing Random Number Distributions


This notebook is for looking at distributions of numbers using the scipy.stats generator, to use mainly as parameter inputs for randomized hyperparameter searches.

See https://en.wikipedia.org/wiki/List_of_probability_distributions for good illustrations of these and other distributions - this script is mainly for tweaking the parameters and seeing what happens.

```python
import numpy as np
import pandas as pd
import scipy.stats as ss
%matplotlib inline
```

## Negative Binomial Distribution

Positive integers, useful for min_samples_split, min_samples_leaf, etc.

```python
x = pd.DataFrame(ss.nbinom(n=10,p=0.5,loc=1).rvs(100000))
print(x.describe(percentiles=[0.01,0.05,0.1,0.25,0.75,0.9,0.95,0.99]))
x.hist(bins=50,range=[0,50],figsize=[12,6])
```

## Beta Distribution

Continuous in [0,1] range, useful for max_features parameters, learning rates, etc.

```python
x = pd.DataFrame(ss.beta(a=3,b=3).rvs(100000))
print(x.describe(percentiles=[0.01,0.05,0.1,0.25,0.75,0.9,0.95,0.99]))
x.hist(bins=50,figsize=[12,6],range=[0,1])
```

## Beta Prime Distribution

Continuous in [0,inf] range, useful for things like C values in logistic regression

```python
x = pd.DataFrame(ss.betaprime(a=2,b=2).rvs(100000))
print(x.describe(percentiles=[0.01,0.05,0.1,0.25,0.75,0.9,0.95,0.99]))
x.hist(bins=50,figsize=[12,6],range=[0,10])
```
